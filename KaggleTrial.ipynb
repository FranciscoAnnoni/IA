```python
import numpy as np
import tensorflow as tf

#Generadores para sets de entrenamiento y pruebas
data_gen_entrenamiento = tf.keras.utils.image_dataset_from_directory(
    '/content/images/ACCchildrenheart',
    validation_split=0.2,
    subset="training",
    color_mode='grayscale',
    seed=123,
    image_size=(250, 250),
    shuffle=True,
    batch_size=35,
    labels="inferred"
)

data_gen_validacion = tf.keras.utils.image_dataset_from_directory('/content/images/ACCchildrenheart',
  validation_split=0.2,
  subset="validation",
  color_mode='grayscale',
  seed=123,
  image_size=(250,250),
  shuffle=True,
  batch_size=35,
  labels="inferred"
)
                                           

#Imprimir 10 imagenes del generador de entrenamiento
```

```python
import matplotlib.pyplot as plt
class_names = data_gen_entrenamiento.class_names
print(class_names)
plt.figure(figsize=(5, 5))
for images, labels in data_gen_entrenamiento.take(1):
    for i in range(10):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"), cmap='gray')
        plt.xlabel(class_names[labels[i]])
plt.show()
```

```python
modeloCNN2 = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(250,250, 1)),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),

  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(250, activation='relu'),
  tf.keras.layers.Dense(3, activation='softmax')
])
modeloCNN2.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

historial=modeloCNN2.fit(data_gen_entrenamiento,epochs=1,batch_size=32,validation_data=data_gen_validacion)
finalModel = modeloCNN2.save("finalModel")
```

```python
import matplotlib.pyplot as plt
class_names = data_gen_entrenamiento.class_names
print(class_names)
plt.figure(figsize=(5, 5))
for images, labels in data_gen_entrenamiento.take(1):
    for i in range(10):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"), cmap='gray')
        plt.xlabel(class_names[labels[i]])
plt.show()
```

```python
modeloCNN2 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(250,250, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(250, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')  # Dos neuronas de salida con activaci√≥n softmax
])
modeloCNN2.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',  # Funci√≥n de p√©rdida adecuada para clasificaci√≥n binaria
                   metrics=['accuracy'])

historial1 = modeloCNN2.fit(data_gen_entrenamiento, epochs=1, batch_size=32, validation_data=data_gen_entrenamiento)
finalModel1 = modeloCNN2.save("finalModel1")
```
